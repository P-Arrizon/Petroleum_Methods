```python
import matplotlib.pyplot as plt
```

```python
def BulkleyLeverett():
    Vp = 50 * 300 * 1000 * 0.23 / 5.614 #bbls
    swi = 0.25
    q = 800  #bbl/day
    break_t = Vp /(q*6.30) #break through time in days
    times = [50, break_t, 200, 500, 750, 1500]
    dfwp = [Vp/(i * q) for i in times] #is the inverse of Td
    Np_lst = [q*50*Vp/614535, 0.15873015873015872*Vp] #inputting the values taht we can easily calculate
    Swp = [.39,0.45,0.475,0.52]
    fwp = [0.76, 0.91,0.94,0.96]
    for i in range(0,len(fwp)):
        sw_avg = Swp[i] + (1 - fwp[i])/dfwp[i+1]
        ans = sw_avg - swi
        Np_lst.append(ans *Vp)

    plt.xlim(0,1750), plt.xlabel("time (days)"), plt.title("Np vs. Time")
    plt.ylim(0,200000), plt.ylabel("Barrles of oil produced")
    plt.plot(121.92,0.15873015873015872*Vp, 'ro', label = "break through time")
    plt.legend(),plt.grid()
    plt.plot(times,Np_lst)
    plt.show()

    # Code below plots SW vs Xd 

    td = 0.1
    dfw_lst = [0,0.05,0.1,0.3,0.7,1.5,3.2]
    sw_lst = [0.7,0.65,0.6,0.55,0.5,0.45,0.4]
    xd_lst = [i *td for i in dfw_lst]
    plt.xlabel("Xd")
    plt.ylabel("Water Sat. (Sw)"), plt.ylim(0,1)
    plt.plot(xd_lst,sw_lst), plt.grid(), plt.title("Sw vs. Xd")
    plt.show()

```

```python
BulkleyLeverett()
```
